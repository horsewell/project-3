```python
# Get modules
import pandas as pd
import plotly.graph_objects as go
import yfinance as yf
```

```python
#Importing data
tickers = ["META", "AMZN", "NFLX", "AAPL"]
df = yf.download(
    tickers = tickers,
    period = "10y",
    interval = "1wk",
    group_by = "column")
df.head()
```

```python
fig = go.Figure()
```

```python
fig.add_trace(go.Scatter(x = aaplc, y = aaplv, mode = "markers", name = "APPL"))
fig.add_trace(go.Scatter(x = amznc, y = amznv, mode = "markers", name = "AMZN"))
fig.add_trace(go.Scatter(x = metac, y = metav, mode = "markers", name = "META"))
fig.add_trace(go.Scatter(x = nflxc, y = nflxv, mode = "markers", name = "NFLX"))


fig.update_layout(
    xaxis_title="Stock Price (USD)", yaxis_title="Stock Volume (log)", title = "Stock Prices vs Stock Volume of MANA",
    width = 800,
    height = 600,
    updatemenus=[go.layout.Updatemenu(
        active = 0,
        x = 1.04,
        xanchor = "left",
        y = 0.3,
        yanchor = "top",
        buttons=list([
                dict(
                    label="ALL",
                    method="update",
                    args=[{"visible":[True, True, True, True]},
                          {"title":"All"}]),
                dict(
                    args=[{"visible":[False,False,True,False]},
                          {"title":"META"}],
                    label="META",
                    method="update"
                ),
                dict(
                    args=[{"visible":[True,False,False,False]},
                          {"title":"APPL"}],
                    label="APPL",
                    method="update"
                ),
                dict(
                    args=[{"visible":[False,False,False,True]},
                          {"title":"NFLX"}],
                    label="NFLX",
                    method="update"
                ),
                dict(
                    args=[{"visible":[False,True,False,False]},
                          {"title":"AMZN"}],
                    label="AMZN",
                    method="update"
                )
            ])
)
    ]
        )
fig.show()
```
